# Matplotlib

O Matplotlib é uma biblioteca de plotagem em Python amplamente utilizada para criar visualizações de dados de alta qualidade. É uma ferramenta poderosa e flexível que permite aos desenvolvedores e cientistas de dados gerar gráficos de forma fácil e eficiente.

## Principais Características do Matplotlib:

**Versatilidade**: Com o Matplotlib, é possível criar uma ampla variedade de gráficos, incluindo gráficos de linha, gráficos de dispersão, gráficos de barra, histogramas, gráficos de pizza, e muitos outros tipos de visualizações. Além disso, ele suporta gráficos 2D e 3D.

**Integração com NumPy**: O Matplotlib funciona perfeitamente com NumPy, uma biblioteca fundamental para computação científica em Python, o que facilita a manipulação e visualização de grandes conjuntos de dados.

**Personalização**: Oferece um alto grau de personalização para gráficos. É possível ajustar cores, estilos de linha, fontes, e outros elementos gráficos para criar visualizações que atendam exatamente às necessidades do usuário.

**Interface Simples e Interface Orientada a Objetos**: O Matplotlib possui duas principais abordagens para criar gráficos: a interface de pyplot, que é semelhante ao MATLAB e fácil de usar para iniciantes, e a interface orientada a objetos, que oferece mais controle e flexibilidade para usuários avançados.

**Compatibilidade**: O Matplotlib é compatível com muitas outras bibliotecas do ecossistema Python, como Pandas para análise de dados e Seaborn para visualizações estatísticas aprimoradas.

# Introdução a Matplotlib: Primeiros Gráficos, Figuras, Titulos e Rótulos


**Primeiros Gráficos**


```python
import matplotlib.pyplot as plt

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Criando um gráfico de linha
plt.plot(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Meu Primeiro Gráfico')
plt.show()
```

**Figuras, Títulos e Rótulos**

```python
import matplotlib.pyplot as plt

# Criando uma nova figura
# fig representa a figura em que o gráfico será desenhado
# ax representa o eixo (ou os eixos) dentro da figura onde o gráfico é plotado
fig, ax = plt.subplots()

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Criando um gráfico de linha
ax.plot(x, y)
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('Gráfico Personalizado')
plt.show()
```

Ajustar o layout da figura

```python
import matplotlib.pyplot as plt

# Criando uma nova figura
fig, ax = plt.subplots()

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Criando um gráfico de linha
ax.plot(x, y)
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('Gráfico Personalizado')

# Ajustar o layout
fig.tight_layout()

plt.show()
```

Adicionar uma anotação na figura:



```python
import matplotlib.pyplot as plt

# Criando uma nova figura
fig, ax = plt.subplots()

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Criando um gráfico de linha
ax.plot(x, y)
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('Gráfico Personalizado')

# Adicionar uma anotação
fig.text(0.5, -0.05, 'Nota: Este é um gráfico de exemplo', ha='center', fontsize=12)

plt.show()
```

```python
# Criando um gráfico de linha com anotação
plt.plot(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico com Anotação')

# Adicionando anotação
plt.annotate('Ponto Interessante', xy=(3, 5), xytext=(4, 6),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

**Tipos de Gráficos**

Além de gráficos de linha, o Matplotlib suporta diversos tipos de gráficos:

Gráfico de Barras

```python
# Dados
categorias = ['A', 'B', 'C', 'D']
valores = [4, 7, 1, 8]

# Criando um gráfico de barras
plt.bar(categorias, valores)
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Gráfico de Barras')
plt.show()
```

```python
# Dados
categorias = ['A', 'B', 'C', 'D']
valores = [4, 7, 1, 8]

# Criando um gráfico de barras
plt.barh(categorias, valores)
plt.ylabel('Categorias')
plt.xlabel('Valores')
plt.title('Gráfico de Barras')
plt.show()
```

Gráfico de Dispersão


```python
# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Criando um gráfico de dispersão
plt.scatter(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Dispersão')
plt.show()
```

# Personalização Matplotlib: Subplots, Paletas de Cores e Salvando Gráficos

Subplots


```python
# Criando subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Dados
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 9, 16, 25]

# Primeiro subplot
ax1.plot(x, y1, 'b-')  # linha azul
ax1.set_title('Gráfico 1')
ax1.set_xlabel('Eixo X')
ax1.set_ylabel('Eixo Y')

# Segundo subplot
ax2.plot(x, y2, 'r-')  # linha vermelha
ax2.set_title('Gráfico 2')
ax2.set_xlabel('Eixo X')
ax2.set_ylabel('Eixo Y')

plt.tight_layout()
plt.show()
```

Paletas de Cores


```python
# Usando uma paleta de cores
plt.style.use('ggplot')

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Criando um gráfico de linha
plt.plot(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico com Estilo ggplot')
plt.show()
```

```python
# Usando uma paleta de cores
plt.style.use('seaborn-v0_8-bright')

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Criando um gráfico de linha
plt.plot(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico com Estilo seaborn-v0_8-bright')
plt.show()
```

Configurações de Eixos


```python
# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Criando um gráfico de linha com configuração de eixos
plt.plot(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico com Configurações de Eixos')

# Configurando os eixos
plt.xlim(0, 6)
plt.ylim(0, 12)
plt.xticks([1, 2, 3, 4, 5], ['A', 'B', 'C', 'D', 'E'])
plt.show()
```

Legendas

```python
# Dados
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 9, 16, 25]

# Criando um gráfico de linha com legendas
plt.plot(x, y1, label='Série 1')
plt.plot(x, y2, label='Série 2')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico com Legendas')
plt.legend()
plt.show()
```

Interatividade


```python
from mpl_toolkits.mplot3d import Axes3D

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
z = [1, 4, 9, 16, 25]

# Criando um gráfico 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')
ax.set_title('Gráfico 3D')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
z = [1, 4, 9, 16, 25]

# Criando um gráfico 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')

# Usar fig.suptitle para o título geral
fig.suptitle('Gráfico 3D')

# Ajustar layout para garantir que as legendas dos eixos sejam legíveis
plt.subplots_adjust(left=-0.5, right=0.9, top=0.9, bottom=0.1)

plt.show()
```
